```python
from typing import Any
from langchain.document_loaders import(
    PyPDFLoader, TextLoader, 
    UnstructuredWordDocumentLoader,
    UnstructuredEPubLoader
)
```

```python
class EpubReader(UnstructuredEPubLoader):
    def __init__(self, file_path: str| list[str], **kwargs:Any):
        super().__init__(file_path,**kwargs,mode="elements", strategy="fast")
```

```python
class DocumentLoaderException(Exception):
    pass
```

```python
class DocumentLoader(object):
    supported_extensions={
        ".pdf": PyPDFLoader,
        ".txt": TextLoader,
        ".epub": EpubReader,
        ".docx": UnstructuredWordDocumentLoader,
        ".doc": UnstructuredWordDocumentLoader
    }
```

```python
import logging
import pathlib
from langchain.schema import Document

def load_document(temp_filepath: str) -> list[Document]:
    ext = pathlib.Path(temp_filepath).suffix
    loader = DocumentLoader.supported_extensions.get(ext)
    if not loader:
        raise DocumentLoaderException(
            f"Invalid extension type {ext}"
        )
    loader = loader(temp_filepath)
    docs = loader.load()
    logging.info(docs)
    return docs
```

```python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.vectorstores import DocArrayInMemorySearch
from langchain.schema import Document, BaseRetriever
```

```python
def configure_retriever(docs: list[Document])->BaseRetriever:
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=1500, chunk_overlap=200)
    splits = text_splitter.split_documents(docs)
    embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
    vectordb = DocArrayInMemorySearch.from_documents(splits,embeddings)
    return vectordb.as_retriever(search_type="mmr", search_kwargs={"k":2, "fetch_k":4})
```

```python
Retriever looking up doc by MMR = Maximum Marginal Relevance
2 options for retriever:
    1. Similarity Search 
    2. Maximum Marginal Relevance:
        Re ranking of documents during retrieval to get results that cover different perspective or point of view from the doc retrieved so FutureWarning
        
Contextual Compression:
        Retrieve doc are compressed.
        Irrevant doc are filtered.
        
        1. LLMChainExtractor:
            Extract from each returned document only the relevant content
        2. LLMChainFilter:
            Filter only relevant documents rather than content from the document
        3. EmbeddingsFilter
            Similarity filter vased on document and query terms of embeddings
```

```python
from langchain.chains import ConversationalRetrievalChain
from langchain.chains.base import Chain
from langchain.chat_models import ChatOpenAI
from langchain.memory import ConversationBufferMemory
```

```python
def configure_chain(retriever: BaseRetriever)->Chain:
    memory = ConversationBufferMemory(
        memory_key="chat_history", return_messages=True
    )
    llm = ChatOpenAI(
        model_name="gpt-3.5-turbo", temperature=0, streaming=True
    )
    return ConversationalRetrievalChain.from_llm(
        llm, retriever=retriever,memory=memory,verbose=True, max_tokens_limit=4000
    )
```

```python
import os
import tempfile
```

```python
def configure_qa_chain(uploaded_file):
    docs=[]
    temp_dir = tempfile.TemporaryDirectory()
    for file in uploaded_file:
        temp_filepath = os.path.join(temp_dir.name, file.name)
        with open(temp_filepath, "wb") as f:
            f.write(file.getvalue())
        docs.extend(load_document(temp_filepath))
    retriever = configure_retriever(docs=docs)
    return configure_chain(retriever=retriever)
```

```python
import streamlit as st
from langchain.callbacks import StreamlitCallbackHandler

st.set_page_config(page_title="Langchain with Documents")
st.title("Langchain with Documents")

uploaded_file = st.sidebar.file_uploader(label="Upload a file", type=list(DocumentLoader.supported_extensions.keys()), accept_multiple_files=True)
if not uploaded_file:
    st.info("Please upload document to continue")
    st.stop()
qa_chain = configure_qa_chain(uploaded_file=uploaded_file)
assistant = st.chat_message("assistant")
user_query = st.chat_input(placeholder="Ask me anything")

if user_query:
    stream_handler = StreamlitCallbackHandler()
    response = qa_chain.run(user_query, callbacks=[stream_handler])
    st.markdown(response)
```
